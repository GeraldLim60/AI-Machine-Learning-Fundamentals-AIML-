## Section 1: Basics of Matplotlib

### 1.1 - Understanding figures and axes

In Matplotlib, a figure can be thought of as a container that holds all elements of your plot. Axes are what we think of as 'plots'. A figure can contain many Axes, but a given Axes object can only be in one Figure. The Axes contains two (or three in the case of 3D) Axis objects.

**Example 1: Creating a figure and axes**

````python
import matplotlib.pyplot as plt

fig = plt.figure()  # an empty figure with no axes
ax = fig.add_subplot(111)  # a figure with a single axes
````

**Example 2: Creating a figure with two axes (plots)**

````python
fig = plt.figure()
ax1 = fig.add_subplot(121)  # subplot with 1 row, 2 columns, and current axis is 1st subplot axes
ax2 = fig.add_subplot(122)  # subplot with 1 row, 2 columns, and current axis is 2nd subplot axes
````

**Example 3: Creating a figure with 2x2 grid of Axes**

````python
fig, axs = plt.subplots(2, 2)  # a figure with a 2x2 grid of Axes
````

**Example 4: Plotting on an Axes**

````python
fig = plt.figure()
ax = fig.add_subplot(111)
ax.plot([1, 2, 3, 4], [10, 20, 25, 30], color='lightblue', linewidth=3)
ax.scatter([0.3, 3.8, 1.2, 2.5], [11, 25, 9, 26], color='darkgreen', marker='^')
ax.set_xlim(0.5, 4.5)
plt.show()
````

**Example 5: Plotting on a figure with two Axes**

````python
x = range(0, 10)
y = range(10, 0, -1)

fig = plt.figure()
ax1 = fig.add_subplot(121)
ax1.plot(x, y)
ax2 = fig.add_subplot(122)
ax2.plot(y, x)
plt.show()
````

### 1.2 - Creating subplots

Subplots are groups of smaller axes that can exist together within a single figure. These can be helpful when you want to compare different views of your data side by side.

**Example 1: Creating a Figure With 2x2 Grid of Subplots**

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

fig, axs = plt.subplots(2, 2)
axs[0, 0].plot(x, y)
axs[0, 0].set_title('Axis [0, 0]')
axs[0, 1].plot(x, y, 'tab:orange')
axs[0, 1].set_title('Axis [0, 1]')
axs[1, 0].plot(x, -y, 'tab:green')
axs[1, 0].set_title('Axis [1, 0]')
axs[1, 1].plot(x, -y, 'tab:red')
axs[1, 1].set_title('Axis [1, 1]')

for ax in axs.flat:
    ax.set(xlabel='x-label', ylabel='y-label')

for ax in axs.flat:
    ax.label_outer()
````

**Example 2: Creating 2 Subplots With Different Sizes**

````python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(5, 3), dpi=100)

ax1.plot(x, y)
ax1.set_title('Subplot 1')

ax2.plot(x, -y)
ax2.set_title('Subplot 2')
````

**Example 3: Sharing the X-axis in Subplots**

````python
fig, (ax1, ax2) = plt.subplots(2, sharex=True)
fig.suptitle('Sharing X axis')
ax1.plot(x, y)
ax2.plot(x + 1, -y)
````

**Example 4: Creating Subplots With Different Scales**

````python
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('time (s)')
ax1.set_ylabel('exp', color=color)
ax1.plot(x, y, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('sin', color=color)  # we already handled the x-label with ax1
ax2.plot(x, y, color=color)
ax2.tick_params(axis='y', labelcolor=color)
````

### 1.3 - Customising plots (colours, markers, line styles)

The visual appearance of the plot can be modified by changing the colours, markers, and line styles.

**Example 1: Changing Line Colour**

The colour of the line can be modified using the `color` keyword.

````python
plt.plot(x, y, color='red')  # red line
plt.show()
````

**Example 2: Changing Line Style**

The style of the line can be modified using the `linestyle` keyword.

````python
plt.plot(x, y, linestyle='dashed')  # dashed line
plt.show()
````

**Example 3: Changing Marker Style**

The marker style can be changed using the `marker` keyword.

````python
plt.plot(x, y, marker='o')  # circle marker
plt.show()
````

**Example 4: Combining Line Colour, Line Style, and Marker Style**

The line colour, line style, and marker style can be combined in a single `plot` function.

````python
plt.plot(x, y, color='green', linestyle='dotted', marker='o')  # green dotted line with circle marker
plt.show()
````

**Example 5: Customising Multiple Lines in the Same Plot**

Each line in the plot can be customised individually.

````python
plt.plot(x, y, color='blue', linestyle='solid', marker='o')  # blue solid line with circle marker
plt.plot(x, y/2, color='red', linestyle='dashed', marker='x')  # red dashed line with cross marker
plt.show()
````

## Section 2: Different Types of Plots

### 2.1 - Line Plot

Line plot is a type of plot which displays information as a series of data points called 'markers' connected by straight line segments. It is a basic type of chart common in many fields.

**Example 1: Basic Line Plot**

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.show()
````

**Example 2: Line Plot with Multiple Lines**

````python
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

plt.plot(x, y1)
plt.plot(x, y2)
plt.show()
````

**Example 3: Line Plot with Title and Labels**

````python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title('Sine Wave')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
````

**Example 4: Line Plot with Grid Lines**

````python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.title('Sine Wave')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.grid(True)
plt.show()
````

**Example 5: Line Plot with Custom Line Style and Markers**

````python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y, color='red', linestyle='dashed', marker='o')
plt.title('Sine Wave')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
````

### 2.2 - Scatter Plot

A scatter plot uses dots to represent values for two different numeric variables. The position of each dot on the horizontal and vertical axis indicates values for an individual data point. Scatter plots are used to observe relationships between variables.

**Example 1: Basic Scatter Plot**

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.scatter(x, y)
plt.show()
````

**Example 2: Scatter Plot with Multiple Data Sets**

````python
x1 = np.linspace(0, 10, 30)
y1 = np.sin(x1)

x2 = np.linspace(0, 10, 30)
y2 = np.cos(x2)

plt.scatter(x1, y1)
plt.scatter(x2, y2)
plt.show()
````

**Example 3: Scatter Plot with Title and Labels**

````python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.scatter(x, y)
plt.title('Scatter Plot of Sine Wave')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
````

**Example 4: Scatter Plot with Custom Markers**

````python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.scatter(x, y, marker='x')
plt.title('Scatter Plot with X Markers')
plt.xlabel('Time')
plt.ylabel('Amplitude')
plt.show()
````

**Example 5: Scatter Plot with Different Sizes and Colors for Points**

````python
x = np.random.rand(50)
y = np.random.rand(50)

colors = np.random.rand(50)
area = np.pi * (15 * np.random.rand(50))**2

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.title('Scatter Plot with Different Sizes and Colors')
plt.show()
````

### 2.3 - Histogram

A histogram is an approximate representation of the distribution of numerical data. It is an estimate of the probability distribution of a continuous variable. To construct a histogram, the first step is to "bin" the range of values — that is, divide the entire range of values into a series of intervals — and then count how many values fall into each interval.

Here are some examples of creating histograms in Python using Matplotlib:

**Example 1: Basic Histogram**

````python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)

plt.hist(data)
plt.show()
````

**Example 2: Histogram with Defined Number of Bins**

````python
data = np.random.randn(1000)

plt.hist(data, bins=20)
plt.title('Histogram with 20 Bins')
plt.show()
````

**Example 3: Histogram with Normalized Values**

````python
data = np.random.randn(1000)

plt.hist(data, bins=20, density=True)
plt.title('Normalized Histogram')
plt.show()
````

**Example 4: Histogram with Cumulative Values**

````python
data = np.random.randn(1000)

plt.hist(data, bins=20, cumulative=True)
plt.title('Cumulative Histogram')
plt.show()
````

**Example 5: Histogram of Several Data Sets**

````python
data1 = np.random.randn(1000)
data2 = np.random.randn(1000)

plt.hist(data1, bins=20, alpha=0.5, label='data 1')
plt.hist(data2, bins=20, alpha=0.5, label='data 2')
plt.title('Histogram of Two Data Sets')
plt.legend(loc='upper right')
plt.show()
````

### 2.4 - Bar Plot

Bar plots are used to display and compare the frequency, count, total or average of different categories of data. They can be horizontal or vertical. They are good if your data is categorical.

**Example 1: Basic Bar Plot**

````python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 10]

plt.bar(categories, values)
plt.show()
````

**Example 2: Bar Plot with Title and Labels**

````python
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 10]

plt.bar(categories, values)
plt.title('Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
````

**Example 3: Horizontal Bar Plot**

````python
categories = ['A', 'B', 'C', 'D']
values = [10, 15, 7, 10]

plt.barh(categories, values)
plt.title('Horizontal Bar Plot')
plt.xlabel('Values')
plt.ylabel('Categories')
plt.show()
````

**Example 4: Stacked Bar Plot**

````python
import numpy as np

categories = ['A', 'B', 'C', 'D']
values1 = [10, 15, 7, 10]
values2 = [5, 7, 6, 8]

plt.bar(categories, values1, label='Group 1')
plt.bar(categories, values2, bottom=values1, label='Group 2')
plt.title('Stacked Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.legend()
plt.show()
````

**Example 5: Grouped Bar Plot**

````python
import numpy as np

categories = ['A', 'B', 'C', 'D']
values1 = [10, 15, 7, 10]
values2 = [5, 7, 6, 8]

barWidth = 0.35
r1 = np.arange(len(categories))
r2 = [x + barWidth for x in r1]

plt.bar(r1, values1, color='b', width=barWidth, edgecolor='grey', label='Group 1')
plt.bar(r2, values2, color='c', width=barWidth, edgecolor='grey', label='Group 2')

plt.xlabel('Categories', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(categories))], categories)
plt.legend()
plt.show()
````

### 2.5 - Area Plot

Area plots are useful for comparing two or more quantities. They allow us to visually represent the magnitude of trends rather than just the trend directions.

**Example 1: Basic Area Plot**

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0.0, 2, 0.01)
y1 = np.sin(2 * np.pi * x)
y2 = 1.2 * np.sin(4 * np.pi * x)

fig, ax = plt.subplots()
ax.fill_between(x, y1, y2)
plt.show()
````

**Example 2: Stacked Area Plot**

````python
x= range(1,6)
y1= [1,4,6,8,9]
y2= [2,2,7,10,12]
y3= [2,8,5,10,6]

plt.stackplot(x,y1,y2,y3)
plt.show()
````

**Example 3: Area Plot With Labels**

````python
x= range(1,6)
y1= [1,4,6,8,9]
y2= [2,2,7,10,12]
y3= [2,8,5,10,6]

plt.stackplot(x,y1,y2,y3, labels=['Y1','Y2','Y3'])
plt.legend(loc='upper left')
plt.show()
````

**Example 4: Area Plot With Different Colors**

````python
x= range(1,6)
y1= [1,4,6,8,9]
y2= [2,2,7,10,12]
y3= [2,8,5,10,6]

plt.stackplot(x,y1,y2,y3, colors=['m','c','r'])
plt.show()
````

**Example 5: Area Plot With Transparency**

````python
x= range(1,6)
y1= [1,4,6,8,9]
y2= [2,2,7,10,12]
y3= [2,8,5,10,6]

plt.stackplot(x,y1,y2,y3, colors=['m','c','r'], alpha=0.5 )
plt.show()
````

### 2.6 - Pie Chart

A pie chart (or a circle chart) is a circular statistical graphic, which is divided into slices to illustrate numerical proportion. The arc length of each slice is proportional to the quantity it represents.

**Example 1: Basic Pie Chart**

````python
import matplotlib.pyplot as plt

sizes = [20, 30, 50]

plt.pie(sizes)
plt.show()
````

**Example 2: Pie Chart with Labels**

````python
labels = ['Group 1', 'Group 2', 'Group 3']
sizes = [20, 30, 50]

plt.pie(sizes, labels=labels)
plt.show()
````

**Example 3: Pie Chart with Percentage**

````python
import matplotlib.pyplot as plt

labels = ['Group 1', 'Group 2', 'Group 3']
sizes = [20, 30, 50]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')
plt.show()
````

**Example 4: Pie Chart with Exploded Slice**

````python
labels = ['Group 1', 'Group 2', 'Group 3']
sizes = [20, 30, 50]
explode = (0, 0.1, 0)  # only "explode" the 2nd slice

plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%')
plt.show()
````

**Example 5: Donut Pie Chart**

````python
labels = ['Group 1', 'Group 2', 'Group 3']
sizes = [20, 30, 50]
explode = (0, 0, 0.1)  # only "explode" the 3rd slice

plt.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', wedgeprops=dict(width=0.3))
plt.show()
````

### 2.7 - Error bars

Error bars are graphical representations of the variability of data and used on graphs to indicate the error, or uncertainty in a reported measurement. They give a general idea of how precise a measurement is, or conversely, how far from the reported value the true (error free) value might be.

Here are some examples of creating error bars in Python using Matplotlib:

**Example 1: Basic Error Bars**

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x, y, yerr=dy, fmt='.k')
plt.show()
````

**Example 2: Error Bars with Different Format**

````python
x = np.linspace(0, 10, 50)
dy = 0.8
y = np.sin(x) + dy * np.random.randn(50)

plt.errorbar(x, y, yerr=dy, fmt='o', color='black',
             ecolor='lightgray', elinewidth=3, capsize=0)
plt.show()
````

**Example 3: Horizontal Error Bars**

````python
x = np.linspace(0, 10, 50)
dx = 0.5
y = np.sin(x)

plt.errorbar(x, y, xerr=dx, fmt='o', color='black',
             ecolor='lightgray', elinewidth=3, capsize=0)
plt.show()
````

**Example 4: Error Bars with Fill Between**

````python
x = np.linspace(0, 10, 100)
dy = 0.1
y = np.sin(x) + dy * np.random.randn(100)

plt.plot(x, y, 'k-')
plt.fill_between(x, y - dy, y + dy, color='gray', alpha=0.2)
plt.show()
````

**Example 5: Error Bars with Asymmetric Sizes**

````python
x = np.linspace(0, 10, 10)
dy = 0.1
y = np.sin(x) + dy * np.random.randn(10)

plt.errorbar(x, y, yerr=[0.1, 0.2, 0.1, 0.3, 0.2, 0.4, 0.1, 0.5, 0.3, 0.6], fmt='o')
plt.show()
````

### 3.2 - Setting x and y axis limits

The x and y axis limits of a plot can be set using the `xlim()` and `ylim()` functions, respectively. These functions take in two arguments: the lower limit and the upper limit.

**Example 1: Setting x and y axis limits**

````python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4])
plt.xlim(0, 5)
plt.ylim(-1, 5)
plt.show()
````

**Example 2: Setting x and y axis limits with Negative Values**

````python
import matplotlib.pyplot as plt

plt.plot([-10, -5, 0, 5, 10])
plt.xlim(-15, 15)
plt.ylim(-15, 15)
plt.show()
````

**Example 3: Setting x and y axis limits for Subplots**

````python
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(2)
ax1.plot([1, 2, 3, 4])
ax1.set_xlim(0, 5)
ax1.set_ylim(-1, 5)
ax2.plot([-10, -5, 0, 5, 10])
ax2.set_xlim(-15, 15)
ax2.set_ylim(-15, 15)
plt.show()
````

**Example 4: Setting x and y axis limits Dynamically**

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-10, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.xlim(x.min()*1.1, x.max()*1.1)
plt.ylim(y.min()*1.1, y.max()*1.1)
plt.show()
````

### 3.3 - Annotating the plot

Annotating involves adding labels or text to the plot at specific points to provide additional information. These can be used to highlight specific observations or patterns in the plot.

**Example 1: Basic Annotation**

You can add a text to a specific point on the plot using the `annotate()` function.

````python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4])
plt.annotate('this is a point', (2, 3))
plt.show()
````

**Example 2: Arrow Annotation**

You can add an arrow to the text annotation pointing to a specific point using the `arrowprops` parameter.

````python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4])
plt.annotate('this is a point', (2, 3), textcoords="offset points", xytext=(-20,-30), arrowprops=dict(arrowstyle='->'))
plt.show()
````

**Example 3: Annotating with Text and Arrow**

You can also customize the annotation by changing the style of the arrow and the text.

````python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4])
plt.annotate('this is a point', (2, 3), textcoords="offset points", xytext=(-20,-30),
             arrowprops=dict(facecolor='black', arrowstyle='->'), fontsize=12, color='red')
plt.show()
````

### 3.4 - Using style sheets

Style sheets in matplotlib are a feature that allows for a lot of creativity and customization. They can change the look and feel of a plot, and they can be used to make your plots more visually appealing and professional looking.

**Example 1: Using a Style Sheet**

The simplest way to use a style sheet is to call `style.use()` with the name of the style sheet you want to use.

````python
import matplotlib.pyplot as plt
import matplotlib.style as style

style.use('ggplot')

plt.plot([1, 2, 3, 4])
plt.show()
````

**Example 2: Listing All Available Style Sheets**

You can list all available style sheets using `style.available`.

````python
import matplotlib.style as style

print(style.available)
````

**Example 3: Combining Multiple Style Sheets**

You can combine multiple style sheets using `style.use()`.

````python
import matplotlib.pyplot as plt
import matplotlib.style as style

style.use(['dark_background', 'bmh'])

plt.plot([1, 2, 3, 4])
plt.show()
````

**Example 4: Temporarily Using a Style Sheet**

If you want to use a style sheet temporarily, you can use the `style.context()` function to create a context manager.

````python
import matplotlib.pyplot as plt
import matplotlib.style as style

with style.context('dark_background'):
    plt.plot([1, 2, 3, 4])
    plt.show()
````

## Section 4: Saving Plots

Saving plots allows for the reuse of the generated plots in presentations, documents, and websites. It also maintains the quality of the plot when used in different mediums.

Here are some examples of saving plots in Python using Matplotlib:

You can save your plots to a file using the `savefig()` function.

**Example 1: Saving Plot as PNG**

````python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4])
plt.savefig('plot.png')
````

**Example 2: Saving Plot as PDF**

````python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4])
plt.savefig('plot.pdf')
````

**Example 3: Saving Plot with Specified DPI**

The quality of the saved plot can be specified using the `dpi` parameter, which stands for dots per inch.

````python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4])
plt.savefig('plot.png', dpi=300)
````

**Example 4: Saving Entire Figure**

Sometimes part of your figure may be cut off when you save it to a file, the `bbox_inches` parameter can be used to specify which part of the figure you want to save.

````python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4])
plt.savefig('plot.png', bbox_inches='tight')
````

**Example 5: Saving Figure with Transparent Background**

By default, the background of the figure will be opaque. However, you can make it translucent or transparent using the `transparent` parameter.

````python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3, 4])
plt.savefig('plot.png', transparent=True)
````
