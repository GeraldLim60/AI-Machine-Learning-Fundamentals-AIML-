# 03.02 - Data Visualisation using Seaborn

## Introduction

This Jupyter Notebook is designed to introduce the basics of data visualization using Seaborn, a Python data visualization library based on matplotlib. The topics we will cover in this notebook include:

1. **Basic Plotting with Seaborn** - We will start by introducing the fundamental types of plots that can be created using Seaborn, such as line plots, bar plots, histograms, and scatter plots. Each of these types of plots can be used to represent different types of data and relationships between variables.
2. **Advanced Seaborn Plots** - After mastering the basics, we will move on to more complex types of plots, such as box and whisker plots, violin plots, heatmaps, and pairplots. These types of plots can reveal more nuanced patterns and relationships in your data.

## Section 1: Basic Plotting with Seaborn

### 1.1 - Line plots

Line plots are often used to visualize a trend in data over intervals of time - a time series.

**Example 1: Basic Line Plot**

> **NOTE:** Remember to install `seaborn` package before proceeding in your virtual environment.

````python
import seaborn as sns

# Load the example flights dataset and convert to long-form
flights_long = sns.load_dataset("flights")

print(flights_long)

# Draw each year in the dataset as a separate line
sns.lineplot(data=flights_long, x = 'year', y = 'passengers')
````

````python
# Load the example tips dataset
tips = sns.load_dataset("tips")

print(tips)

# Draw a nested lineplot to show daily tips by time
sns.lineplot(x="day", y="tip", hue="time", data=tips)
````

**Example 3: Line Plot with Several Variables**

````python
# Load the example iris dataset
iris = sns.load_dataset("iris")

print(iris)

# Draw a line plot with several variables
sns.lineplot(x="sepal_width", y="sepal_length", hue="species", style="species", data=iris)
````

**Example 4: Line Plot with Customized Markers**

````python
# Draw a line plot with customized markers
sns.lineplot(x="sepal_width", y="sepal_length", hue="species", style="species", markers=True, dashes=False, data=iris)
````

### 1.2 - Bar Plots

Bar plots are often used to compare the quantity, frequency or other measure (such as mean) for different categories or groups.

**Example 1: Basic Bar Plot**

````python
# Load the example tips dataset
tips = sns.load_dataset("tips")

# Draw a set of vertical bar plots grouped by a categorical variable
sns.barplot(x="day", y="total_bill", data=tips)
````

**Example 2: Horizontal Bar Plot**

````python
# Draw a set of horizontal bar plots grouped by a categorical variable
sns.barplot(x="total_bill", y="day", data=tips)
````

**Example 3: Bar Plot with Hue**

````python
# Draw a set of bar plots grouped by a categorical variable and split by another categorical variable
sns.barplot(x="day", y="total_bill", hue="sex", data=tips)
````

**Example 4: Bar Plot with Custom Colours**

````python
# Draw a set of bar plots with a custom color palette
sns.barplot(x="day", y="total_bill", hue="sex", palette="Blues", data=tips)
````

**Example 5: Bar Plot with Error Bars**

````python
# Draw a set of bar plots with error bars to represent the variability in the data
sns.barplot(x="day", y="total_bill", hue="sex", errorbar="sd", capsize=.2, data=tips)
````

### 1.3 - Histograms

Histograms are used to plot the distribution of a numeric variable. The x-axis represents bins or intervals, and the y-axis represents the frequency.

**Example 1: Basic Histogram**

````python
# Load the example tips dataset
tips = sns.load_dataset("tips")

# Draw a histogram of a column
sns.histplot(data=tips, x="total_bill")
````

**Example 2: Histogram with Specified Bin Size**

````python
# Draw a histogram of a column with specified bin size
sns.histplot(data=tips, x="total_bill", bins=10)
````

**Example 3: Histogram for Each Category**

````python
# Draw separate histograms for each category
sns.histplot(data=tips, x="total_bill", hue="day")
````

**Example 4: Histogram with Density Curve**

````python
# Draw a histogram with a density curve
sns.histplot(data=tips, x="total_bill", kde=True)
````

**Example 5: Overlapping Histograms**

````python
# Load another example dataset
penguins = sns.load_dataset("penguins")

# Draw overlapping histograms
sns.histplot(data=penguins, x="flipper_length_mm", hue="species", multiple="stack")
````

## Section 2: Advanced Seaborn Plots

### 2.1 - Box and Whisker plots

Box and whisker plots are used to show the distribution of quantitative data in a way that facilitates comparisons between variables or across levels of a categorical variable.

**Example 1: Basic Box Plot**

````python
# Load the example tips dataset
tips = sns.load_dataset("tips")

# Draw a simple box plot
sns.boxplot(x=tips["total_bill"])
````

**Example 2: Horizontal Box Plot**

````python
# Draw a horizontal box plot
sns.boxplot(y=tips["total_bill"])
````

**Example 3: Grouped Box Plot**

````python
# Draw a box plot for each day of the week
sns.boxplot(x="day", y="total_bill", data=tips)
````

**Example 4: Box Plot with Hue**

````python
# Draw a set of box plots split by a categorical variable
sns.boxplot(x="day", y="total_bill", hue="smoker", data=tips)
````

**Example 2: Horizontal Violin Plot**

````python
# Draw a horizontal violin plot
sns.violinplot(y=tips["total_bill"])
````

**Example 3: Grouped Violin Plot**

````python
# Draw a violin plot for each day of the week
sns.violinplot(x="day", y="total_bill", data=tips)
````

**Example 4: Violin Plot with Hue**

````python
# Draw a set of violin plots split by a categorical variable
sns.violinplot(x="day", y="total_bill", hue="smoker", data=tips)
````

### 2.3 - Heatmaps

Heatmaps are used to visualize data in 2-dimensional form. The individual values within the matrix are represented by colors.

Here, we will use the predefined "iris" dataset from Python's seaborn library. This dataset contains measurements for 150 iris flowers from three different species.

**Example 1: Basic Heatmap**

````python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# load the iris dataset
iris = sns.load_dataset('iris')

# initiates the label encoder
le = LabelEncoder()

# apply label encoding to the 'species' column
iris['species'] = le.fit_transform(iris['species'])

# Compute correlation matrix
corr = iris.corr()

# Create a heatmap
sns.heatmap(corr)
````

**Example 2: Heatmap with Labels**

````python
# Create a heatmap with labels
sns.heatmap(corr, annot=True, fmt=".2f")
````

**Example 3: Heatmap with Custom Color Map**

````python
# Create a heatmap with a custom color map
sns.heatmap(corr, cmap="YlGnBu", annot=True, fmt=".2f")
````

**Example 4: Heatmap with Diverging Color Map**

````python
# Create a heatmap with a diverging color map
sns.heatmap(corr, cmap="coolwarm", center=0, annot=True, fmt=".2f")
````

**Example 5: Heatmap with Customized Color Bar**

````python
# Create a heatmap with a customized color bar
sns.heatmap(corr, cmap="coolwarm", center=0, cbar_kws={"label": "Correlation Coefficient"}, annot=True, fmt=".2f")
````

### 2.4 - Pairplots

Pairplots are used to quickly visualize and analyze the relationship between pairs of features in a dataset. It creates a grid of Axes where each variable in your data is shared across the y-axes across a single row and the x-axes across a single column.

**Example 1: Basic Pairplot**

````python
# Load the example iris dataset
iris = sns.load_dataset("iris")

# Create a basic pairplot
sns.pairplot(iris)
````

**Example 2: Pairplot with Hue**

````python
# Create a pairplot with different categories represented by different colors
sns.pairplot(iris, hue="species")
````

**Example 3: Pairplot with Specific Variables**

````python
# Create a pairplot with specific variables
sns.pairplot(iris, vars=["sepal_width", "sepal_length"])
````

**Example 4: Pairplot with Regression Lines**

````python
# Create a pairplot with regression lines
sns.pairplot(iris, kind="reg")
````

**Example 5: Pairplot with Custom Marker Styles**

````python
# Create a pairplot with different marker styles for each species
sns.pairplot(iris, hue="species", markers=["o", "s", "D"])
````

## Section 3: Customizing Seaborn Plots

### 3.1 - Adding titles and labels

Adding titles and labels to your Seaborn plots is an essential step to provide context for your visualizations. The `matplotlib` library, which Seaborn is built upon, offers functions like `title()`, `xlabel()`, and `ylabel()`.

**Example 1: Adding a Title to Your Plot**

````python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset("tips")
sns.histplot(data=tips, x="total_bill")
plt.title('Total Bill Distribution')
````

**Example 2: Adding X and Y Labels**

````python
sns.histplot(data=tips, x="total_bill")
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
````

**Example 3: Customizing the Title and Labels**

You can also customize the font size, color, and other properties of your text.

````python
sns.histplot(data=tips, x="total_bill")
plt.title('Total Bill Distribution', fontsize=20, color='blue')
plt.xlabel('Total Bill', fontsize=15, color='green')
plt.ylabel('Frequency', fontsize=15, color='red')
````

**Example 4: Using Latex in Title and Labels**

You can use Latex in your titles and labels to write mathematical expressions.

````python
sns.histplot(data=tips, x="total_bill")
plt.title('Total Bill Distribution ($\\mu=19.78$)', fontsize=20)
plt.show()
````

### 3.2 - Changing color palettes

You can change the color palette of your plot in Seaborn. Seaborn offers several default color palettes and also allows you to create your own custom color palettes.

**Example 1: Default Color Palette**

````python
import seaborn as sns

tips = sns.load_dataset("tips")

sns.scatterplot(x="total_bill", y="tip", data=tips)
````

**Example 2: Using a Different Default Color Palette**

````python
sns.set_palette("husl")
sns.scatterplot(x="total_bill", y="tip", data=tips)
````

**Example 3: Using a Custom Color Palette**

````python
my_palette = ["#4B0082", "#7F00FF", "#E0B0FF"]
sns.set_palette(my_palette)
sns.scatterplot(x="total_bill", y="tip", data=tips)
````

**Example 4: Using a Sequential Color Palette**

A sequential color palette is suitable for data that ranges from relatively low or uninteresting values to relatively high or interesting ones.

````python
sns.set_palette("Blues")
sns.scatterplot(x="total_bill", y="tip", data=tips)
````

**Example 5: Using a Diverging Color Palette**

A diverging color palette is suitable for data that deviates around a middle value, especially when the data has both a logical zero point and meaningful negative and positive values.

````python
sns.set_palette("coolwarm")
sns.scatterplot(x="total_bill", y="tip", data=tips)
````

### 3.3 - Adjusting plot size and shape

You can adjust the size and shape of your plot in Seaborn. The size and shape of a plot can significantly affect the story that the plot tells.

**Example 1: Creating a Plot of a Specified Size**

````python
plt.figure(figsize=(8, 6))
sns.scatterplot(x="total_bill", y="tip", data=tips)
````

**Example 2: Creating a Square Plot**

````python
plt.figure(figsize=(6, 6))
sns.scatterplot(x="total_bill", y="tip", data=tips)
````

**Example 3: Creating a Plot with a Specified Aspect Ratio**

````python
plt.figure(figsize=(8, 4))
sns.scatterplot(x="total_bill", y="tip", data=tips)
````

**Example 4: Adjusting the Size of a Plot After It's Created**

````python
plot = sns.scatterplot(x="total_bill", y="tip", data=tips)
plot.figure.set_size_inches(10, 8)
````

**Example 5: Creating a Plot with a Specified DPI (Dots Per Inch)**

````python
plt.figure(figsize=(8, 6), dpi=120)
sns.scatterplot(x="total_bill", y="tip", data=tips)
````
